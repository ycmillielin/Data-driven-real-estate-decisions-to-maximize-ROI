```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import censusdata
import requests
```

**Florida**

```python
data=pd.read_excel('FloridaPropertyData.xlsx')
data=data.rename(columns={"PROPERTY_LATITUDE": "lat",'PROPERTY_LONGITUDE':'lon'})
data = data[data['lat'].notna()]
data = data[data['lon'].notna()]
data
```

```python
url = "https://geo.fcc.gov/api/census/block/find"
data=data.assign(
    block=data.apply(
        lambda r: requests.get(
            url, params={"latitude": r["lat"], "longitude": r["lon"], "format": "json"}
        ).json()["Block"]["FIPS"],
        axis=1,
    )
)
```

```python
data['block']=data['block'].astype(str)
data['tract'] = [x[:-4] for x in data['block']]
data
```

```python
data.to_csv('FloridaMappedData2.csv')
```
